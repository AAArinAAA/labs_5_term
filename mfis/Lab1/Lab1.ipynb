```python
def get_mod_string(text):
    mod_list = []
    for char in text:
        mod = alfabet.get(char) % 31
        mod_list.append(mod)
    return mod_list
```

```python
def find_k(mod_list):
    for i in range (1, 32):
        key_string = ''
        print (f"Проверка ключа {i}")
        for mod in mod_list:
            new_mod = (mod + i ) % 32
            for key in alfabet:
                if alfabet[key] == new_mod:
                    key_string = key_string + key
        print (key_string)
```

```python
def code_with_key(text, key):
    output_str = ''
    char_array = list(text)
    for char in char_array:
        new_mod = (alfabet.get(char) - key ) % 32
        for key_d in alfabet:
            if alfabet[key_d] == new_mod:
                output_str = output_str + key_d
    print (output_str)
```
