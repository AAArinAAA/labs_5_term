```python
def is_prime(x):
    for i in range(2, (x//2)+1):
        if x % i == 0:
            return False
    return True
```

```python
is_prime(4)
```

```python
a_list = []

for i in range(0, len(alfabet)):
    if is_prime(i):
        a_list.append(i)
    else: 
        if math.gcd(i, len(alfabet)) == 1:
            a_list.append(i)

a_inv_list = []

for a in a_list:
    if a != 0 and a%2 != 0:
        print( pow(a, -1, len(alfabet)))
        a_inv_list.append(a)
                     
        
```

```python
def affine_cipher_code(text, a, b):
    if a not in a_list:
        print ('Choose other a coefficient')
    else:
        cipher_text = ''
        char_array = list(text)
        for char in char_array:
            new_char = (a * alfabet.get(char) + b) % len(alfabet)
            for key_d in alfabet:
                if alfabet[key_d] == new_char:
                    cipher_text = cipher_text + key_d
        print (cipher_text)
```

```python
affine_cipher_code(text, a, b)
```
